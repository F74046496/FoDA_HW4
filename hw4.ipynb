```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import linear_model, metrics, model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
```

### Read Data

```python
train_name = 'dota2Train.csv'
data_train = pd.read_csv(train_name, sep=',')
test_name = 'dota2Test.csv'
data_test = pd.read_csv(test_name, sep=',')
data_train.head(10)
```

### Data Preparation

```python
data_train = data_train.drop(['id'], axis=1)
data_test = data_test.drop(['id'], axis=1)
data_train.head(10)
```

```python
x_train = data_train.loc[:, data_train.columns != 'win']
y_train = data_train['win']
x_test = data_test.loc[:, data_test.columns != 'win']
y_test = data_test['win']
#data_train['1'].value_counts()
```

```python
data_train2 = data_train[data_train.modee == 2]
#data_test2 = data_test.drop(['mode', 'type'], axis=1)
data_train2.head(10)
data_train2['modee'].value_counts()
```

```python
data_train3 = data_train2[data_train2.type == 2]
#data_test2 = data_test.drop(['mode', 'type'], axis=1)
data_train3.head(10)
len(data_train3)
```

```python
data_train3 = data_train3.drop(['modee', 'type'], axis=1)
data_test3 = data_test.drop(['modee', 'type'], axis=1)
#data_train3 = data_train3
#data_test3 = data_test
data_train3.head()
x_train3 = data_train3.loc[:, data_train3.columns != 'win']
y_train3 = data_train3['win']
x_test3 = data_test3.loc[:, data_test3.columns != 'win']
y_test3 = data_test3['win']
```

### Logistic Regression

```python
clf3 = LogisticRegression(solver='lbfgs').fit(x_train3, y_train3)
clf3.fit(x_train3, y_train3)

lr_acc3 = metrics.accuracy_score(y_test3, clf3.predict(x_test3))
print('LogisticRegression accuracy: %s' % lr_acc3)
```
